```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True  
        if not swapped:
            break

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def quick_sort(arr):
    if len(arr) <= 1:
        return arr 
    else:
        pivot = arr[0]
        less_than_pivot = [x for x in arr[1:] if x <= pivot]
        greater_than_pivot = [x for x in arr[1:] if x > pivot]
        
        return quick_sort(less_than_pivot) + [pivot] + quick_sort(greater_than_pivot)
```

```python
import random;
import matplotlib.pyplot as plt
import time;

def measure_sort_performance(sort_functions, start_size=50, end_size=2000, step=1):
    performance_data = {sort_function.__name__: [] for sort_function in sort_functions}

    for i in range(start_size, end_size, step):
        base_array = [random.randint(1, 2000) for _ in range(i)]

        for sort_function in sort_functions:
            array_to_sort = base_array.copy()
            start_time = time.time()
            sort_function(array_to_sort)
            end_time = time.time()

            execution_time = end_time - start_time
            performance_data[sort_function.__name__].append((i, execution_time))

    for sort_function in sort_functions:
        x_values = [pair[0] for pair in performance_data[sort_function.__name__]]
        y_values = [pair[1] for pair in performance_data[sort_function.__name__]]

        plt.plot(x_values, y_values, linestyle='-', linewidth=0.7, label=f"{sort_function.__name__}")

    plt.xlabel('Number of Elements (n)')
    plt.ylabel('Execution Time (seconds)')
    plt.title('Performance of Sorting Algorithms')
    plt.legend()
    plt.grid(True)
    plt.show()

measure_sort_performance([bubble_sort, insertion_sort,quick_sort])
```
